```python
#files used which dont have overlapping wavelengths
file1='SN2016ccj_uv_20160514_1.dat'
file2='SN2016ccj_uv_20160514_2.dat'

optfile1='SN2016ccj_opt_20160514_1.dat'

optwave1=[]
optflux1=[]
wave1=[]
flux1=[]
wave2=[]
flux2=[]

f = open(optfile1, 'r')
for line in f:
    #print(line)
    content = line.split()
    optwave1.append(float(content[0]))
    optflux1.append(float(content[1]))

f = open(file1, 'r')
for line in f:
    #print(line)
    content = line.split()
    wave1.append(float(content[0]))
    flux1.append(float(content[1]))

f = open(file2, 'r')
for line in f:
    #print(line)
    content = line.split()
    wave2.append(float(content[0]))
    flux2.append(float(content[1]))

```

```python
plt.plot(wave1, flux1, 'c-', label = "1", alpha = 1, linewidth = .7)
plt.plot(optwave1, optflux1, '-', label = "opt", alpha = 1, linewidth = .7)
plt.plot(wave2, flux2, '-', label = "2", alpha = 1, linewidth = .7)
#plt.yscale('log')
#plt.xlim([2900,3200])
plt.ylim([0,2*10**(-15)])
#plt.ylim([10**(-16),10**(-14)])

#plt.legend()
#axes = plt.axes()
#axes.set_xlim([1600,5600])
#axes.set_ylim([10**(-16),10**(-14)])
plt.show()
```
